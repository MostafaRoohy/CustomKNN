### Distance Functions and Sort Neighbors Function

```python
def GetLorentzianDistance(i, j):

    theTestingSample = featuresNumPy[i]
    theTrainSample = featuresNumPy[j]

    theirDistance = np.sum((np.log((np.abs(theTestingSample-theTrainSample) + 1))))

    return (theirDistance)
#


def GetEuclidianDistance(i, j):

    theTestingSample = featuresNumPy[i]
    theTrainSample = featuresNumPy[j]

    theirDistance = np.sqrt(np.sum(np.power((np.abs(theTestingSample-theTrainSample)),2)))

    return (theirDistance)
#



def GetMostafaDistance(i, j):

    theTestingSample = featuresNumPy[i]
    theTrainSample = featuresNumPy[j]

    theirDistance = abs((np.sum(theTestingSample))-(np.sum(theTrainSample)))

    return (theirDistance)
#


def GetWeightedLorentzianDistance(i, j, weight):

    theTestingSample = featuresNumPy[i]
    theTrainSample = featuresNumPy[j]

    theirLorentDistance = 0.0

    for k in range(0, theTestingSample.size):

        theirLorentDistance += math.log(1 + abs((theTestingSample[k]*weight)-(theTrainSample[k]*weight)))
    #


    return (theirLorentDistance)
#


def GetWeightedEuclidianDistance(i, j, weight):

    theTestingSample = featuresNumPy[i]
    theTrainSample = featuresNumPy[j]

    theirLorentDistance = 0.0

    for k in range(0, theTestingSample.size):

        theirLorentDistance += pow(((theTestingSample[k]*weight)-(theTrainSample[k]*weight)), 2)
    #

    theirLorentDistance = math.sqrt(theirLorentDistance)


    return (theirLorentDistance)
#


def GetWeightedMostafaDistance(i, j, weight):

    theTestingSample = featuresNumPy[i]
    theTrainSample = featuresNumPy[j]

    theirDistance = abs(((np.sum(theTestingSample)))-((np.sum(theTrainSample))))

    return (theirDistance*weight)
#



def Sort(sub_li):

    sub_li.sort(key=lambda x: x[0])
    return sub_li
#
```

### Mostafa KNN

```python
def ThisTradeKNN(i, whichDistanceMetric, trainingSetCount, nearK, nearVotesNeeded, farK, farVotesNeeded):


    theVotes = []

    weightDiscount = 1/(trainingSetCount)
    weight = 1

    for j in range(i-5, i-5-trainingSetCount, -1):

        theDistance = 0.0


        if (whichDistanceMetric == 1):
            theDistance = GetLorentzianDistance(i, j)
        #
        elif (whichDistanceMetric == 2):
            theDistance = GetEuclidianDistance(i, j)
        #
        elif (whichDistanceMetric == 3):
            theDistance = GetMostafaDistance(i, j)
        #
        elif (whichDistanceMetric == 4):
            theDistance = GetWeightedLorentzianDistance(i, j, weight)
        #
        elif (whichDistanceMetric == 5):
            theDistance = GetWeightedEuclidianDistance(i, j, weight)
        #
        elif (whichDistanceMetric == 6):
            theDistance = GetWeightedMostafaDistance(i, j, weight)
        #
            

        theVotes.append([theDistance, labelsNumPy[j]])


        weight-=weightDiscount
    #
            

    Sort(theVotes)


    nearVotes = [sublist[-1] for sublist in theVotes[:nearK]]
    farVotes  = [sublist[-1] for sublist in theVotes[-farK:]]

    theNearResult  = 1 if (nearVotes.count(1) >= nearVotesNeeded) else 0
    theFarResult   = 1 if (farVotes.count(0)  >= farVotesNeeded) else 0


    return (theNearResult, theFarResult)
#          (firstUnsortedVote, allVotesResult, theNearResult, theFarResult)
```

```python
def ProcessSubSet(startCalcAt, endCalcAt, trainingSetCount, nearK, nearVotesNeeded, farK, farVotesNeeded, useMartingale, slSize, rrSize, startCapital,  whichDistanceMetric, howHokm):


    mojaz_true_count, mojaz_false_count, mamnoo_true_count, mamnoo_false_count = 0, 0, 0, 0

    lostTradesSteps = 0
    winTradesSteps = 0


    # //////////////////////////////////////////////////////////////////////////////////////////////
    for i in range(startCalcAt, endCalcAt):


        if (not(useMartingale)):
            winTradesSteps = 0
            lostTradesSteps = 0
        #
            

        (theNearResult, theFarResult) = ThisTradeKNN(i, whichDistanceMetric, trainingSetCount, nearK, nearVotesNeeded, farK, farVotesNeeded)
       

        hokm = False

        if (howHokm==1):
            hokm =  (theNearResult==1)
        #        
        elif (howHokm==2):
            hokm =  (theFarResult==0)
        #
        elif (howHokm==3):
            hokm =  ((theNearResult==1)  and  (theFarResult==0))
        #        
        elif (howHokm==4):
            hokm =  ((theNearResult==1)  or  (theFarResult==0))
        #        
        

        tpSize = slSize * rrSize

        
        if (hokm):
            
            if (labelsNumPy[i] == 1):

                mojaz_true_count += 1

                
                if(lostTradesSteps>0):
                    startCapital += (tpSize*lostTradesSteps)
                #
                else:
                    startCapital += tpSize
                #


                winTradesSteps += 1
                lostTradesSteps = 0
            #
            elif (labelsNumPy[i] == 0):

                mojaz_false_count += 1


                if(lostTradesSteps>0):
                    startCapital -= (slSize*lostTradesSteps)
                #
                else:
                    startCapital -= slSize
                #


                winTradesSteps = 0
                lostTradesSteps += 1
            #
        #
        else:
            if (labelsNumPy[i] == 1):

                mamnoo_true_count += 1
            #
            elif (labelsNumPy[i] == 0):

                mamnoo_false_count += 1
            #
        #
    # ////////////////////////////////////////////////////////////////////////////////////////////// 




    tradesExecutedCount = (mojaz_true_count+mojaz_false_count)
    winRate = (math.floor(mojaz_true_count / tradesExecutedCount*100)) if (tradesExecutedCount > 0) else 0


    return (tradesExecutedCount, winRate, startCapital)
#          (tradesExecutedCount, winRate, startCapital)
```

```python
def Cumulater(startCalcAt, endCalcAt, trainingSetCount, nearK, nearVotesNeeded, farK, farVotesNeeded, useMartingale, slSize, rrSize, subsetsStartingCapital, subsetsCumulativeProfit, subsetsCounter, bestWinRate, worstWinRate, subsetWinratesSums, whichDistanceMetric, howHokm):


    tradesExecutedCount, winRate, profit = ProcessSubSet(startCalcAt, endCalcAt, trainingSetCount, nearK, nearVotesNeeded, farK, farVotesNeeded, useMartingale, slSize, rrSize, subsetsStartingCapital, whichDistanceMetric, howHokm)


    pureProfit = (profit-subsetsStartingCapital)
    subsetsCumulativeProfit += pureProfit
    subsetsCounter += 1


    subsetWinratesSums += winRate
    if (winRate > bestWinRate):
        bestWinRate = winRate
    if (winRate < worstWinRate):
        worstWinRate = winRate
    #
        
    outputColor = Fore.WHITE
    if (tradesExecutedCount==0):
        outputColor = Fore.BLACK
    elif (pureProfit>0):
        outputColor = Fore.GREEN
    elif (pureProfit<=0):
        outputColor = Fore.RED
    #
        
        
    print(outputColor, tradeOpened_yearNumber[startCalcAt], "_", tradeOpened_monthNumber[startCalcAt], ":\t", startCalcAt, "\t--> ", endCalcAt, "\t", "Executed Trades:", tradesExecutedCount, "\t", "WR:", winRate, "%", "\t", "PnL:", (pureProfit))
#
```
