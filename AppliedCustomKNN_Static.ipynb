### Distance Functions and Sort Neighbors Function

```python
def GetLorentzianDistance(i, j):

    theTestingSample = featuresNumPy[i]
    theTrainSample = featuresNumPy[j]

    theirDistance = np.sum((np.log((np.abs(theTestingSample-theTrainSample) + 1))))

    return (theirDistance)
#


def GetEuclidianDistance(i, j):

    theTestingSample = featuresNumPy[i]
    theTrainSample = featuresNumPy[j]

    theirDistance = np.sqrt(np.sum(np.power((np.abs(theTestingSample-theTrainSample)),2)))

    return (theirDistance)
#


def GetEuclidianDistanceForThisFeatures(i, j, features):

    features = np.float64(features)

    theTestingSample = features[i]
    theTrainSample = features[j]

    theirDistance = np.sqrt(np.sum(np.power((np.abs(theTestingSample-theTrainSample)),2)))

    return (theirDistance)
#


def GetMostafaDistance(i, j):

    theTestingSample = featuresNumPy[i]
    theTrainSample = featuresNumPy[j]

    theirDistance = abs((np.sum(theTestingSample))-(np.sum(theTrainSample)))

    return (theirDistance)
#


def GetWeightedLorentzianDistance(i, j, weight):

    theTestingSample = featuresNumPy[i]
    theTrainSample = featuresNumPy[j]

    theirLorentDistance = 0.0

    for k in range(0, theTestingSample.size):

        theirLorentDistance += math.log(1 + abs((theTestingSample[k]*weight)-(theTrainSample[k]*weight)))
    #


    return (theirLorentDistance)
#


def GetWeightedEuclidianDistance(i, j, weight):

    theTestingSample = featuresNumPy[i]
    theTrainSample = featuresNumPy[j]

    theirLorentDistance = 0.0

    for k in range(0, theTestingSample.size):

        theirLorentDistance += pow(((theTestingSample[k]*weight)-(theTrainSample[k]*weight)), 2)
    #

    theirLorentDistance = math.sqrt(theirLorentDistance)


    return (theirLorentDistance)
#


def GetWeightedMostafaDistance(i, j, weight):

    theTestingSample = featuresNumPy[i]
    theTrainSample = featuresNumPy[j]

    theirDistance = abs(((np.sum(theTestingSample)))-((np.sum(theTrainSample))))

    return (theirDistance*weight)
#



def Sort(sub_li):

    sub_li.sort(key=lambda x: x[0])
    return sub_li
#
```

### KNN

```python
def ThisTradeKNN(i, whichDistanceMetric, trainingSetCount, nearK, nearVotesNeeded, farK, farVotesNeeded):

    theVotes = []
    consequtiveNessVotes = []
    TrendtoPBVotes   = []
    TrendtoBOVotes   = []
    PBtoBOVotes      = []
    PBmaximumVotes   = []
    

    weightDiscount = 1/(trainingSetCount)
    weight = 1

    for j in range(i-5, i-5-trainingSetCount, -1):

        theDistance = 0.0


        if (whichDistanceMetric == 1):
            theDistance = GetLorentzianDistance(i, j)
        #
        elif (whichDistanceMetric == 2):

            theDistance = GetEuclidianDistance(i, j)

            consequtiveNessDistance = GetEuclidianDistanceForThisFeatures(i, j, consequtiveNess)
            TrendtoPBDistance       = GetEuclidianDistanceForThisFeatures(i, j, TrendtoPB)
            TrendtoBODistance       = GetEuclidianDistanceForThisFeatures(i, j, TrendtoBO)
            PBtoBODistance          = GetEuclidianDistanceForThisFeatures(i, j, PBtoBO)
            PBmaximumDistance       = GetEuclidianDistanceForThisFeatures(i, j, PBmaximum)

        #
        elif (whichDistanceMetric == 3):
            theDistance = GetMostafaDistance(i, j)
        #
        elif (whichDistanceMetric == 4):
            theDistance = GetWeightedLorentzianDistance(i, j, weight)
        #
        elif (whichDistanceMetric == 5):
            theDistance = GetWeightedEuclidianDistance(i, j, weight)
        #
        elif (whichDistanceMetric == 6):
            theDistance = GetWeightedMostafaDistance(i, j, weight)
        #


        theVotes.append([theDistance, labelsNumPy[j]])
        consequtiveNessVotes.append([consequtiveNessDistance, labelsNumPy[j]])
        TrendtoPBVotes.append([TrendtoPBDistance, labelsNumPy[j]])
        TrendtoBOVotes.append([TrendtoBODistance, labelsNumPy[j]])
        PBtoBOVotes.append([PBtoBODistance, labelsNumPy[j]])
        PBmaximumVotes.append([PBmaximumDistance, labelsNumPy[j]])


        weight-=weightDiscount
    #

    Sort(theVotes)
    Sort(consequtiveNessVotes)
    Sort(TrendtoPBVotes)
    Sort(TrendtoBOVotes)
    Sort(PBtoBOVotes)
    Sort(PBmaximumVotes)

  
    allVotes  = [sublist[-1] for sublist in theVotes[:]]
    nearVotes = [sublist[-1] for sublist in theVotes[:nearK]]
    farVotes  = [sublist[-1] for sublist in theVotes[-farK:]]
    allVotesResult = 1 if (allVotes.count(1)  >= (trainingSetCount/2)) else 0
    theNearResult  = 1 if (nearVotes.count(1) >= nearVotesNeeded) else 0
    theFarResult   = 1 if (farVotes.count(1)  >= farVotesNeeded) else 0


    nearconsequtiveNessVotes = [sublist[-1] for sublist in consequtiveNessVotes[:nearK]]
    nearTrendtoPBVotes       = [sublist[-1] for sublist in TrendtoPBVotes[:nearK]]
    nearTrendtoBOVotes       = [sublist[-1] for sublist in TrendtoBOVotes[:nearK]]
    nearPBtoBOVotes          = [sublist[-1] for sublist in PBtoBOVotes[:nearK]]
    nearPBmaximumVotes       = [sublist[-1] for sublist in PBmaximumVotes[:nearK]]


    nearconsequtiveNessResult  = 1 if (nearconsequtiveNessVotes.count(1) >= nearVotesNeeded) else 0
    nearTrendtoPBResult   = 1 if (nearTrendtoPBVotes.count(1) >= nearVotesNeeded) else 0
    nearTrendtoBOResult = 1 if (nearTrendtoBOVotes.count(1) >= nearVotesNeeded) else 0
    nearPBtoBOResult = 1 if (nearPBtoBOVotes.count(1) >= nearVotesNeeded) else 0
    nearPBmaximumResult  = 1 if (nearPBmaximumVotes.count(1) >= nearVotesNeeded) else 0




    return (theNearResult, theFarResult, nearconsequtiveNessResult, nearTrendtoPBResult, nearTrendtoBOResult, nearPBtoBOResult, nearPBmaximumResult)
#          
```

```python
def ProcessSubSet (startCalcAt, endCalcAt, trainingSetCount, nearK, nearVotesNeeded, farK, farVotesNeeded, startCapital, useMartingale, whichDistanceMetric, howHokm):


    mojaz_true_count, mojaz_false_count, mamnoo_true_count, mamnoo_false_count = 0, 0, 0, 0

    lostTradesSteps = 0
    winTradesSteps = 0


    # //////////////////////////////////////////////////////////////////////////////////////////////
    for i in range(startCalcAt, endCalcAt):

        if (not(useMartingale)):
            winTradesSteps = 0
            lostTradesSteps = 0
        #

        (theNearResult, theFarResult, nearconsequtiveNessResult, nearTrendtoPBResult, nearTrendtoBOResult, nearPBtoBOResult, nearPBmaximumResult) = ThisTradeKNN(i, whichDistanceMetric, trainingSetCount, nearK, nearVotesNeeded, farK, farVotesNeeded)
       




        hokm =     (theNearResult  and  (nearTrendtoPBResult    and    nearTrendtoBOResult   and   nearPBtoBOResult)  or  (nearconsequtiveNessResult  or  nearPBmaximumResult))




        # hokm =  theNearResult


        # hokm =  nearTrendtoPBResult


        # hokm =  nearTrendtoBOResult


        # hokm =  nearPBtoBOResult


        # hokm =  nearPBmaximumResult


        # hokm = nearconsequtiveNessResult






        
        if (hokm):
            
            if (labelsNumPy[i] == 1):

                mojaz_true_count += 1

                
                if(lostTradesSteps>0):
                    startCapital += (20*lostTradesSteps)
                #
                else:
                    startCapital += 20
                #


                winTradesSteps += 1
                lostTradesSteps = 0
            #
            elif (labelsNumPy[i] == 0):

                mojaz_false_count += 1


                if(lostTradesSteps>0):
                    startCapital -= (10*lostTradesSteps)
                #
                else:
                    startCapital -= 10
                #


                winTradesSteps = 0
                lostTradesSteps += 1
            #
        #
        else:
            if (labelsNumPy[i] == 1):

                mamnoo_true_count += 1
            #
            elif (labelsNumPy[i] == 0):

                mamnoo_false_count += 1
            #
        #
    # //////////////////////////////////////////////////////////////////////////////////////////////



    tradesExecutedCount = (mojaz_true_count+mojaz_false_count)
    winRate = (math.floor(mojaz_true_count / tradesExecutedCount*100)) if (tradesExecutedCount > 0) else 0


    return (tradesExecutedCount, winRate, startCapital)
#          (tradesExecutedCount, winRate, startCapital)
```

```python
trainingSetCount = 300

nearK           = 3
nearVotesNeeded = 2

farK           = 5
farVotesNeeded = 3


useMartingale        = False  
datasamplesPerSubset = 500

subsetWinratesSums = 0
subsetsCount       = 0

bestWinRate  = 000
worstWinRate = 100

subsetsStartingCapital  = 100
subsetsCumulativeProfit = 0.0

whichDistanceMetric = 2
howHokm             = 5

for subset in range(1000, labelsNumPy.size - datasamplesPerSubset, datasamplesPerSubset):

    
    startCalcAt, endCalcAt = (subset), (subset + datasamplesPerSubset)

    tradesExecutedCount, winRate, profit = ProcessSubSet(startCalcAt, endCalcAt, trainingSetCount, nearK, nearVotesNeeded, farK, farVotesNeeded, subsetsStartingCapital, useMartingale, whichDistanceMetric, howHokm)


    pureProfit = (profit-subsetsStartingCapital)
    subsetsCumulativeProfit += pureProfit


    subsetWinratesSums += winRate
    subsetsCount += 1
    if (winRate > bestWinRate):
        bestWinRate = winRate
    if (winRate < worstWinRate):
        worstWinRate = winRate
    #
        
    outputColor = Fore.WHITE
    if (tradesExecutedCount==0):
        outputColor = Fore.BLACK
    elif (pureProfit>0):
        outputColor = Fore.GREEN
    elif (pureProfit<=0):
        outputColor = Fore.RED
    #
        
    print(outputColor, "DataIndex", subsetsCount, ":\t", startCalcAt, "\t--> ", endCalcAt, "\t", "Total Trades:", tradesExecutedCount, "\t", "WinRate:", winRate, "%", "\t", "Pure Profit:", (pureProfit))
#

print(Fore.WHITE, "\n")
print("Average WinRate  :" , math.floor(subsetWinratesSums/subsetsCount), "%")
print("Best WinRate:"    , bestWinRate, "%")
print("Worst WinRate:"   , worstWinRate, "%")
print("CumulativeProfitt:", subsetsCumulativeProfit, "$")
```
